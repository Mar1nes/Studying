```python
# Exponential distribution

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from scipy import stats

e = stats.expon.rvs(0.2, size=100)
# e = np.sort(e)
# y = np.arange(100)
print(e, sep=' ')

"""
Function stats.expon.rvs generate random 100 values 
from exponential distribution, where mean count of events
is equal 2/period. It means that this values can use for 
modeling 100 intervals (minutes) between service calls where
incoming call mean intencity (lambda) is equal 0.2 per minute.
"""

fig, ax = plt.subplots(figsize=(5, 5))
# ax.plot(e, y, 'r+', scalex=0.1, scaley=1,)
ax.hist(e, bins=20)
ax.set_xlabel('Interval time')
ax.set_ylabel('Count')
plt.title('Histplot 100 intervals with lambda (intencity) = 2 events/period')
plt.show()
```

```python
# Weibull distribution

"""
    It apply when labda is not equal const, for example - 
    the risk of mechanical failure increases over time.
    The Weibull's distribution is extend of the Exponential distribution.
    When shape parametr (betta) > 1, the  failure probabolity increases.
    When betta < 1, the p (probability) is decreases.
    Function stats.weibull_min.rvs returns 100 random vaues (life times) from 
    Weibull's distribution with shape parametr - 1.5 
    and typical life time = 5000. Life time - resource.
"""

from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

w = stats.weibull_min.rvs(1.5, scale=5000, size=100)

fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(w, bins=20)
ax.set_xlabel('Life time')
ax.set_ylabel('Count')
plt.title('Histplot 100 random values betta=1.5, scale(eta)=5000')
plt.show()
```

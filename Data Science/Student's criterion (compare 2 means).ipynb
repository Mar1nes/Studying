```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

# DATA
vid_1 = np.array([84.7, 105, 98.9, 97.9, 108.7, 81.3, 99.4, 89.4, 93.0, 119.3, 99.2, 99.4, 97.1, 112.4, 99.8, 94.7, 114.0, 95.1, 115.5, 111])
vid_2 = np.array([57.2, 68.6, 104.4, 95.1, 89.9, 70.8, 83.5, 60.1, 75.7, 102, 69, 79.6, 68.9, 98.6, 76, 74.8, 56, 55.6, 69.4, 59])


mn1 = np.mean(vid_1).round(2)
mn2 = np.mean(vid_2).round(2)
sd1 = np.std(vid_1).round(2)
sd2 = np.std(vid_2).round(2)
n1 = len(vid_1)
n2 = len(vid_2)
print(f'mn1 = {mn1}')
print(f'mn2 = {mn2}')
print(f'sd1 = {sd1}')
print(f'sd2 = {sd2}')
print(f'n1 = {n1}')
print(f'n2 = {n2}')


df1 = pd.DataFrame(data=vid_1, columns=['Species 1'])
df2 = pd.DataFrame(data=vid_2, columns=['Species 2'])
df = pd.concat([df1, df2], axis=1)

# BOX PLOTS
sns.boxplot(data=df)
plt.title('DNA destraction temperature')
plt.ylabel('Temperature (F)')
plt.show()

# t-test?
ttest = stats.ttest_ind(vid_1, vid_2, equal_var=False)
print(f't = {ttest[0]:.2f}, p = {ttest[1]:.5f}')

if ttest[1] < 0.05:
    print('H0 - False.')
else:
    print('H0 - True')
```

Парный t-тест (сравнение 2х средних выборочных значений)

ФОРМУЛЫ

Дано 2 выборочных средних (Xср1, Xср2), 2 стандартные ошибки (SD1, SD2), 
число наблюдений (n1, n2).

Проверяем 2 гипотезы H0 (средние генеральных совокупностей равны) и H1 
(соответственно не равны).

Если предположить, что гипотеза H0 верна, то распределение значений разницы 
выборочных средних (Xср1 - Xср2) выглядело бы как график снизу слева. По 
центру распределения был бы 0, т.к. гипотеза говорит что значения мю1 и мю2 равны
(мю - средн. значения генеральных совокупностей), следовательно мю1 - мю2 = 0. 

Стандартная ошибка этого распределения (SE) расчитывылась бы по формуле: 
кв. корень из суммы стандартного отклонения 1ой выборки в квадрате (sd1), деленного 
на количество наблюдений 1 выборки (n1) и аналогичного выражения для второй выборки 
(ст. откл. делить на число наблюдений). Обе стандартные ошибки вносят свой вклад.

При достаточно большом количестве наблюдений, можно сказать что наше распределение
(разностей между двумя средними значениями), приняло бы форму t-распределения с числом 
степеней свобод (df), которое расчитывается по формуле n1+n2-2.

Основываясь на всех этих данных, мы можем расчитать насколько далеко наше значение "а"
отклонилось от разности средних генеральных совокупностей, тем самым расчитать вероятность
получения таких или еще более выраженных различий (с отрицательными знаениями) при условии,
что нулевая гипотеза H0 верна. Расчитываем "t" по формуле справа.

Расчитав значение "t" (коэффициент Стьюдента) и зная число степеней свобод (df), мы можем расчитать "p" уровень значимости.

ЗАДАЧА:

Процесс денатурации ДНК представляет собой разрушение водородных связей между двумя цепями этой молекулы и очень сильно зависит от температуры, с которой мы воздействуем на молекулу.

При сравнении двух видов между собой в исследовании были получены следующие различия в средней температуре плавления ДНК|:



Сформулируем гипотезы:

H0: M1 = M2

H1: M1 != M2



коэффициент t = 2,5 означает, что наша разница между выборочными средними отклонилась от разницы средних генеральных совокупностей на 2,5 сигм (стандартных отклонений) вправо.

Теперь расчитаем вероятность получить такие или еще более выраженные отклонения при условии что на самом деле у нас верна нулевая гипотеза (H0). Для этого нам потребуется знать число степеней свобод df = n1 + n2 - 2 = 20 + 20 - 2 = 38.
Вероятность такого события равняется p = 0.0167, т.е. уровень значимости оказался сильно меньше 0,05 (p < 0.05)

Вывод: мы обнаружили статистически значимые различия в средней температуре плавления ДНК у двух видов.






ВАЖНЫЕ ЗАМЕЧАНИЯ:

1) Применяя критерий t-Стьюдента, желательно чтобы дисперсия внутри групп были приблизительно одинаковы (требование гомогенности дисперсии). Чтобы проверить это можно использовать критерий Левене или критерий Фишера.

2) Если объем выборки недостаточно большой (меньше 30), то очень важным требованием является нормальность распределения выборок. Если кол-во наблюдений больше 30, то t-тест справляется нормально даже если распределения этих выборок отличается от нормального, однако это нужно проверять.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

# DATA
vid_1 = np.array([84.7, 105, 98.9, 97.9, 108.7, 81.3, 99.4, 89.4, 93.0, 119.3, 99.2, 99.4, 97.1, 112.4, 99.8, 94.7, 114.0, 95.1, 115.5, 111])
vid_2 = np.array([57.2, 68.6, 104.4, 95.1, 89.9, 70.8, 83.5, 60.1, 75.7, 102, 69, 79.6, 68.9, 98.6, 76, 74.8, 56, 55.6, 69.4, 59])


mn1 = np.mean(vid_1).round(2)
mn2 = np.mean(vid_2).round(2)
sd1 = np.std(vid_1).round(2)
sd2 = np.std(vid_2).round(2)
n1 = len(vid_1)
n2 = len(vid_2)
print(f'mn1 = {mn1}')
print(f'mn2 = {mn2}')
print(f'sd1 = {sd1}')
print(f'sd2 = {sd2}')
print(f'n1 = {n1}')
print(f'n2 = {n2}')


df1 = pd.DataFrame(data=vid_1, columns=['Species 1'])
df2 = pd.DataFrame(data=vid_2, columns=['Species 2'])
df = pd.concat([df1, df2], axis=1)

# BOX PLOTS
sns.boxplot(data=df)
plt.title('DNA destraction temperature')
plt.ylabel('Temperature (F)')
plt.show()

# t-test?
ttest = stats.ttest_ind(vid_1, vid_2, equal_var=False)
print(f't = {ttest[0]:.2f}, p = {ttest[1]:.5f}')

if ttest[1] < 0.05:
    print('H0 - False.')
else:
    print('H0 - True')
```

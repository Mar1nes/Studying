```python
# THE CENTRAL LIMITS THOREM

# THE PROOF

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

df=pd.read_csv('//Users/ivankudravcev/Studying/Data/loans_income.csv')

data = [[] for i in range(3)]
data[0] = df.sample(1000, replace=True)['"x"'].tolist()
for i in range(1000):
    smple2 = df.sample(5, replace=True)
    smple3 = df.sample(20, replace=True)
    data[1].append(np.mean(smple2).values[0])
    data[2].append(np.mean(smple3).values[0])
    
colors=['blue','red','yellow']  
i=0
for d in data:
    sns.histplot(d, color=colors[i], legend=True)
    i+=1
plt.title('red:1000 values | green:1000 means of 5 | blue:1000 means of 20')
plt.show()
```

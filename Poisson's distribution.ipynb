```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

s = np.random.poisson(lam = 2.0, size=100)

plt.plot(s)
plt.title("Poisson's distribution")
plt.xlabel("Time (min)")
plt.ylabel("Count of events")

print("This plot show the Poisson's distribution with lambda=2 and size=100. As model it can be 'Call service' with 2 calls per minute. This code simulate 100 minutes and returns count of calls for each minute.")
```

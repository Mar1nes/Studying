```python
# CORRELATION RESEARCH

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

#  CORRELATION MATRIX FOR FIRST 20 COLUMNS OF DATA FRAME (FOR EXAMPLE)
df = pd.read_csv("//Users/ivankudravcev/Studying/Data/sp500_data.csv")
df1 = df.iloc[:,0:21]
df1.corr()
```

```python
# CORRELATION TABLE FOR FIRST 20 COLUMNS ()
%matplotlib inline
plt.figure(figsize=(15,10), dpi=80)
sns.heatmap(df1.corr(), xticklabels=df1.corr().columns, yticklabels=df1.corr().columns, cmap='RdYlGn', center=0, annot=True)
```

```python
# SEARCHING MAX CORRELATION COEFFICIENT IN CORRELATION MATRIX (ALL DATA)
cor_matrix = df.corr()
cor_matrix = cor_matrix.to_numpy()
cor_matrix -= np.eye(len(cor_matrix))
max_cor = np.max(cor_matrix)
print(f"max cor coefficient = {max_cor.round(2)}")

# SEARCHING POSITION OF MAX CORRELATION COEFFICIENT IN MATRIX
ind = []
for i in range(len(cor_matrix)):
        if np.max(cor_matrix[i]) == max_cor:
            ind.append(i)

print(f'index of max cor in matrix: {ind}')


            
# DRAW PLOT WITH MAX_COR COEFFICIENT (ANSWER: SPY-DIA)
%matplotlib inline
df2 = df.iloc[:, ind[0]:np.min(ind)+10]
plt.figure(figsize=(15,10), dpi=70)
sns.heatmap(df2.corr(), xticklabels=df2.corr().columns, yticklabels=df2.corr().columns, cmap='RdYlGn', center=0, annot=True)
```

```python
# SEARCHING MIN CORRELATION COEFFICIENT IN CORRELATION MATRIX (ALL DATA)
min_cor = np.min(cor_matrix)
print(f'min cor coef = {min_cor.round(2)}')

# SEARCHING POSITION OF MIN CORRELATION COEFFICIENT IN MATRIX

# NUMBER OF STRING (i)
for i in range(len(cor_matrix)):
        if np.min(cor_matrix[i]) == min_cor:
            break

# NUMBER OF ELEMENT IN STRING (j)
j = 0
for n in df[:][i:i+1]:
    if n == min_cor:
        break
    j += 1
print(f'index in string[{i}]: {j}')

# DRAW PLOT AND SEARCH STOCKS (USING PLOT) WITH MAX_COR COEFFICIENT (ANSWER: SPY-DIA)
%matplotlib inline
df3 = df.iloc[j:j+10, i:i+10]
plt.figure(figsize=(15,10), dpi=80)
sns.heatmap(df3.corr(), xticklabels=df3.corr().columns, yticklabels=df3.corr().columns, center=0, cmap='RdYlGn', annot=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

df = pd.read_csv('//Users/ivankudravcev/Studying/Data/kc_tax.csv')
df = df[df['TaxAssessedValue'] < 750000]
df = df[df['SqFtTotLiving'] > 100]
df = df[df['SqFtTotLiving'] < 3500]
df
```

```python
# HEX PLOT
hx = sns.jointplot(data=df, x='SqFtTotLiving', y='TaxAssessedValue', kind='hex')
plt.show()

# KDE PLOT
kde = sns.kdeplot(x=df['SqFtTotLiving'], y=df['TaxAssessedValue'], cmap="Reds", shade=True, thresh=0)
plt.show()
```

```python
# ALL UNIQUE VALUES IN VARIABLE (COLUMN 'ZipCode')
zips = np.array(df['ZipCode'].dropna().unique(), dtype=int)
zips
```

```python
# WHERE IS THE MOST EXPENSIVE SQUARE Ft TOTAL LIVING?

max_tax_ind=np.argmax(df['TaxAssessedValue'])
df.iloc[max_tax_ind]
```

```python
# SCATTER & HEX PLOT AT ZIPCODE 98075 (THE MOST EXPENSIVE PLACE)
sns.jointplot(data=df[df['ZipCode']==98075], x='SqFtTotLiving', y='TaxAssessedValue', kind='scatter')
sns.jointplot(data=df[df['ZipCode']==98075], x='SqFtTotLiving', y='TaxAssessedValue', kind='hex')
```
